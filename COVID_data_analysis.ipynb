```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('covid_19_india.csv')
df
```

```python
df["Active"] = (df["Confirmed"])-(df["Cured"]+df["Deaths"])
df
```

```python
df["State/UnionTerritory"].replace({"Telengana": "Telangana", "Telangana***": "Telangana"}, inplace=True)
df.set_index("State/UnionTerritory")
```

```python
df1=df.groupby("State/UnionTerritory").sum()
df1
```

```python
df4=df1.drop(['Unnamed: 9','Sno'], axis = 1)
df4
```

```python
df2=df1[df1.Active == df1.Active.max()]
df2.drop(['Unnamed: 9','Sno','Confirmed','Cured','Deaths'], axis = 1)
```

```python
df3=df1[df1.Active == df1.Active.min()]
df3.drop(['Unnamed: 9','Sno','Confirmed','Cured','Deaths'], axis = 1)
```

```python
df4["Ratio"] = (df4["Deaths"]/df4["Cured"])
df4
```

```python
df5=df4[df4.Ratio == df4.Ratio.max()]
df5.drop(['Confirmed','Cured','Deaths','Active'], axis = 1)
```

```python
df6=df4[df4.Ratio == df4.Ratio.min()]
df6.drop(['Confirmed','Cured','Deaths','Active'], axis = 1)
```

```python
df_new=df.drop(['Time','Unnamed: 9', 'Sno'],axis=1)
df_new
```

```python
df_delh=df_new.loc[df_new['State/UnionTerritory'].isin(['Delhi'])]
df_delhi=df_delh.drop(['ConfirmedIndianNational','ConfirmedForeignNational','Confirmed'],axis=1)
df_delhi
```

```python
df_maharashtr=df_new.loc[df_new['State/UnionTerritory'].isin(['Maharashtra'])]
df_maharashtra=df_maharashtr.drop(['ConfirmedIndianNational','ConfirmedForeignNational','Confirmed'],axis=1)
df_maharashtra
```

```python
df_delhi_cured=df_delhi['Cured']
df_delhi_date=df_delhi['Date']
df_delhi_deaths=df_delhi['Deaths']
df_delhi_active=df_delhi['Active']
df_maharashtra_cured=df_maharashtra['Cured']
df_maharashtra_date=df_maharashtra['Date']
df_maharashtra_deaths=df_maharashtra['Deaths']
df_maharashtra_active=df_maharashtra['Active']
```

```python
from matplotlib import style
style.use("ggplot")
plt.plot(df_delhi_date,df_delhi_active,'g',label='Delhi', linewidth=2)
plt.plot(df_maharashtra_date,df_maharashtra_active,'r',label='Maharashtra',linewidth=2)
plt.title('Active')
plt.ylabel('Active')
plt.xlabel('Date')
plt.legend()
plt.grid(False, color = 'w')
plt.show()
```

This graph shows that the number of active patients is decreasing in Delhi after a certain date while in Maharashtra it is continuously increasing.

```python
style.use("ggplot")
plt.plot(df_delhi_date,df_delhi_cured,'g',label='Delhi', linewidth=2)
plt.plot(df_maharashtra_date,df_maharashtra_cured,'r',label='Maharashtra',linewidth=2)
plt.title('Cured')
plt.ylabel('Cured')
plt.xlabel('Date')
plt.legend()
plt.grid(False, color = 'w')
plt.show()
```

The rate of patients getting cured with respect to time is more in Maharashtra than in Delhi 

```python
style.use("ggplot")
plt.plot(df_delhi_date,df_delhi_deaths,'g',label='Delhi', linewidth=2)
plt.plot(df_maharashtra_date,df_maharashtra_deaths,'r',label='Maharashtra',linewidth=2)
plt.title('Death')
plt.ylabel('Death')
plt.xlabel('Date')
plt.legend()
plt.grid(False, color = 'w')
plt.show()
```

The death rate is more in Maharashtra than in Delhi . Also, the number of patients dying is also more in Maharashtra than in Delhi
